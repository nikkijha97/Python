```python
import numpy as np
```

```python
##creating numpy
## as an array
a=np.array([[1,2,3],[4,5,6]])
print("a",a)
## with arrange
b=np.arange(1,20,3)
print("b",b)
##with linspace
c=np.linspace(1,20,5)
print("c",c)

##with zeroes
d=np.zeros((3,2))
print("d",d)

##with ones

e=np.ones((6,7))
print("e",e)

```

```python
## MATRIX ##
a=np.array([[1,2,3],[4,5,6]])
print(a)
b=np.array([[8,9,10]])
print(b)
c=np.concatenate((a,b))
print('c',c)
```

```python
##dot product
b=b.transpose()
c=np.dot(a,b)
print(c)
```

***PANDAS***

```python
import pandas as pd
```

```python
##creating dataframe
d = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
## Create a DataFrame df from this dictionary data which has the index labels.

df=pd.DataFrame(d, index= labels)
df
```

```python
##converting data frames  to numpy
arr=df.to_numpy()
arr
```

```python
##converting data frames  to numpy
##converting single or more column to numpy array

arr=df[['age','visits']].to_numpy()
arr
```
