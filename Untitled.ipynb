```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'name': ['nikki','rikki','priyanka','shruti'],
                'salary':[10,30,90,80]})
df1
```

```python
df2=pd.DataFrame({'name': ['nikki','rikki','priyanka','shruti'],
                  'education':['A','B','C','D']})
```

```python
##merging new columns
pd.merge(df1,df2,on='name')
```

```python
con=[
    (df1['salary']<20),
    (df1['salary']>=20) & (df1['salary']<80),
    (df1['salary']<100)
]

res=['low','avg','high']
df1['status']=np.select(con,res)
df1
```
